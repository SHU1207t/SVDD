```python
import MVTecAD_loaddata as mv
import matplotlib.pyplot as plt
import numpy as np
import os
import sys
import csv
import tensorflow as tf
import sklearn
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import OneClassSVM
from sklearn.metrics import roc_auc_score, f1_score, accuracy_score
import cv2
import pandas as pd
from PIL import Image
```

```python
def OCSVM(X, x_test, y, y_test):
    tuned_parameters = [
        {'nu': [2**-3, 2**-4, 2**-5, 2**-6], 'kernel': ['rbf'], 'gamma': [10**-4, 10**-5, 10**-6]},
    ]
    score = 'accuracy'
    clf = GridSearchCV(
        OneClassSVM(),
        tuned_parameters,
        cv=5,
        scoring=score)
    clf.fit(X, y)
    a = clf.decision_function(x_test)
    b = clf.predict(x_test)
    r1 = roc_auc_score(y_test,a)
    r2 = f1_score(y_test,b)
    r3 = accuracy_score(y_test,b)
    print('roc_auc : ',r1)
    print('f1_score : ', r2)
    print('accuracy_score : ', r3)
    print('best_params : ', clf.best_params_)
    result = [r1,r2,r3]
    return result

def fimage_bottle(X):
    data = []
    for i in range(len(X)):
        img = cv2.imread('fakes/bottle/'+str(i)+'.png')
        img = list(img)
        data.append(img)
    data = np.array(data)
    return data

def fimage_cable(X):
    data = []
    for i in range(len(X)):
        img = cv2.imread('fakes/cable/'+str(i)+'.png')
        img = list(img)
        data.append(img)
    data = np.array(data)
    return data     

def fimage_capsule(X):
    data = []
    for i in range(len(X)):
        img = cv2.imread('fakes/capsule/'+str(i)+'.png')
        img = list(img)
        data.append(img)
    data = np.array(data)
    return data     

def fimage_hazelnut(X):
    data = []
    for i in range(len(X)):
        img = cv2.imread('fakes/hazelnut/'+str(i)+'.png')
        img = list(img)
        data.append(img)
    data = np.array(data)
    return data     

def fimage_metal_nut(X):
    data = []
    for i in range(len(X)):
        img = cv2.imread('fakes/metal-nut/'+str(i)+'.png')
        img = list(img)
        data.append(img)
    data = np.array(data)
    return data     

def fimage_pill(X):
    data = []
    for i in range(len(X)):
        img = cv2.imread('fakes/pill/'+str(i)+'.png')
        img = list(img)
        data.append(img)
    data = np.array(data)
    return data     

def fimage_screw(X):
    data = []
    for i in range(len(X)):
        img = cv2.imread('fakes/screw/'+str(i)+'.png')
        img = list(img)
        data.append(img)
    data = np.array(data)
    return data     

def fimage_toothbrush(X):
    data = []
    for i in range(len(X)):
        img = cv2.imread('fakes/toothbrush/'+str(i)+'.png')
        img = list(img)
        data.append(img)
    data = np.array(data)
    return data     

def fimage_transistor(X):
    data = []
    for i in range(len(X)):
        img = cv2.imread('fakes/transistor/'+str(i)+'.png')
        img = list(img)
        data.append(img)
    data = np.array(data)
    return data     

def fimage_zipper(X):
    data = []
    for i in range(len(X)):
        img = cv2.imread('fakes/zipper/'+str(i)+'.png')
        img = list(img)
        data.append(img)
    data = np.array(data)
    return data
res = []
```

```python
X_train, X_test, X_test_good, X_test_error, y_tr, y_tg, y_te, y_ts = mv.bottle()
# X_train, X_test, X_test_good, X_test_error, y_tr, y_tg, y_te, y_ts = mv.cable()
# X_train, X_test, X_test_good, X_test_error, y_tr, y_tg, y_te, y_ts = mv.capsule()
# X_train, X_test, X_test_good, X_test_error, y_tr, y_tg, y_te, y_ts = mv.hazelnut()
# X_train, X_test, X_test_good, X_test_error, y_tr, y_tg, y_te, y_ts = mv.metal_nut()
# X_train, X_test, X_test_good, X_test_error, y_tr, y_tg, y_te, y_ts = mv.pill()
# X_train, X_test, X_test_good, X_test_error, y_tr, y_tg, y_te, y_ts = mv.screw()
# X_train, X_test, X_test_good, X_test_error, y_tr, y_tg, y_te, y_ts = mv.toothbrush()
# X_train, X_test, X_test_good, X_test_error, y_tr, y_tg, y_te, y_ts = mv.transistor()
# X_train, X_test, X_test_good, X_test_error, y_tr, y_tg, y_te, y_ts = mv.zipper()

im = fimage_bottle(X_train)
im = im[:int(len(X_train)*0.125)]/255
x_train = np.concatenate([X_train, im])
y_train = []
for i in range(len(im)):
    y_train.append(-1)
y_train = np.array(y_train)
y_train = y_train.reshape([len(y_train), 1])
labels = np.concatenate([y_tr, y_train])
x_train = np.reshape(x_train, (len(x_train), -1))
x_test = np.reshape(X_test, (len(X_test), -1))
result = OCSVM(x_train, x_test, labels, y_ts)
res.append(result)
```
